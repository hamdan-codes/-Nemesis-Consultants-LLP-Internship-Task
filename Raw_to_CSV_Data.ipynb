# Making CSV from raw text file data

```python
with open('transactions/transactions.txt', 'r') as f: # opening files and reading data as strings
    data = f.readlines()
data
```

```python
df = []
for i in data:
    df.append(i.strip()) # stripping '\n' from the end of the strings
df
```

```python
import pandas as pd
import json
```

```python
data = []
for i in df:
    data.append(json.loads(i)) # Converting to json data to convert to dataframe
```

```python
df = pd.DataFrame.from_dict(data) # Converting to dataframe
df
```

```python
df.to_csv('dataframe.csv', index=False) # Writing dataframe to a csv file to easily access in future
```
